Consider two input arguments, “correct predictions” and “total predictions”. The function should calculate the accuracy value as per the given formula
Accuracy = correct predictions/total predictions × 100. Name the function as calculate_accuracy.


```python
def calculate_accuracy(correct_predictions, total_predictions):
    if total_predictions == 0:
        return 0  # Avoid division by zero
    accuracy = (correct_predictions / total_predictions) * 100
    return accuracy
```

Run the function and obtain the output for the following values
correct predictions = 95
total predictions = 100


```python
correct = 95
total = 100

result = calculate_accuracy(correct, total)

print(f"Accuracy: {result:.2f}%")
```

Modify the code written for previous part above appropriately so that the values for the variables “correct predictions” and “total predictions” can be obtained from a user and then passed as arguments to the function calculate_accuracy to get the output (accuracy).


```python
correct = int(input("Enter the number of correct predictions: "))
total = int(input("Enter the total number of predictions: "))

result = calculate_accuracy(correct, total)

print(f"Accuracy: {result:.2f}%")
```

```python
image_path = '/content/drive/MyDrive/ML Lab/Lenna.png'
```

Import the OpenCV library to read the image file


```python
import cv2

image = cv2.imread(image_path)
```

Convert the input image to RGB


```python
if image is not None:
    print("Image loaded successfully!")
else:
    print("Error: Image not found.")
```

Convert the input image to RGB


```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.title('Lenna Image in RGB')
plt.show()
```

Convert the input image to grayscale


```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.title('Lenna Image in Grayscale')
plt.show()
```

Convert the above grayscale to a binary image (Use a threshold of 127)


```python
_, image_binary = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(image_binary, cmap='gray')
plt.axis('off')
plt.title('Binary Image (Threshold = 127)')
plt.show()
```

Import the matplotlib library and use the subplot command to plot the converted RGB image, grayscale, and binary image in a single figure.


```python
plt.figure(figsize=(12, 4))

# RGB image
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("RGB Image")
plt.axis("off")

# Grayscale image
plt.subplot(1, 3, 2)
plt.imshow(image_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

# Binary image
plt.subplot(1, 3, 3)
plt.imshow(image_binary, cmap='gray')
plt.title("Binary Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
